# EDA

```python
def null_searcher(data):
    count = data.shape[0]
    dic = {}
    for col in data.columns:
        dic[col]= data[col].isna().sum()
    return pd.DataFrame({'columns': list(dic.keys()),
                         'null_ration': [i/count for i in list(dic.values())]}).set_index('columns')
        
```

```python
null_searcher(df)
```

# Data preprocessing

```python
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, mean_squared_error
from sklearn.impute import SimpleImputer
```

```python
train_df.dtypes
```

```python
target = ['target']
categorical = ['neighbourhood_group', 'neighbourhood', 'room_type']
for col in categorical:
    print(col, train_df[col].nunique())
```

```python
# OneHotEncoder
for col in categorical:
    
    ohe = OneHotEncoder()
    train_df[list(ohe.categories_[0])]=ohe.fit_transform(train_df[[col]]).toarray()
    train_df.drop(columns=[col], inplace=True)
```

```python
null_searcher(train_df).sort_values(by=['null_ration'], ascending=False)
```

```python
# SimpleImputer
for col in train_df.columns:
    if train_df[col].isna().sum() > 0:
        si = SimpleImputer()
        train_df[col] = si.fit_transform(train_df[[col]])
        
null_searcher(train_df).null_ration.unique()
```

```python
# StandartScaler

for col in train_df.columns:
    if col != 'target':
        sc = StandardScaler()
        train_df[col] = sc.fit_transform(train_df[[col]])
    else:
        continue
```

```python
train_df.head()
```

# train_test_split and val_split

```python
train, oos = train_test_split(train_df, test_size=0.3, random_state=17)
```

```python
train.shape
```

```python
oos.shape
```

# RidgeCV

```python
ridge = RidgeCV(alphas=(0.001, 0.01, 0.1, 1.0, 10.0, 12, 14), cv=5)
ridge.fit(train.drop(columns='target'), train['target'])
print('best alpha', ridge.alpha_)
print('mean_square_error', mean_squared_error(oos.target, ridge.predict(oos.drop(columns=['target']))))
```

# LassoCV

```python
lasso = LassoCV(n_alphas=500, random_state=17)
lasso.fit(train.drop(columns='target'), train['target'])
print('best alpha', lasso.alpha_)
print('best score')
print('mean_square_error', mean_squared_error(oos.target, lasso.predict(oos.drop(columns=['target']))))
```

# ElasticNetCv

```python
encv = ElasticNetCV(n_alphas=500, random_state=17)
encv.fit(train.drop(columns='target'), train['target'])
print('best alpha', lasso.alpha_)
print('mean_square_error', mean_squared_error(oos.target, encv.predict(oos.drop(columns=['target']))))
```
